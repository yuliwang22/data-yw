```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import scipy
plt.style.use('ggplot')
```

```python
df=pd.read_csv("~/Desktop/datasets/vending_machine_sales.csv")
```

```python
df.head()
df.shape
df.describe()
df.dtypes
```

```python
df.isna().sum()
```

```python
ax = df['Category'].value_counts() \
  .plot(kind='bar', 
        title='Top Categories Sold')
ax.set_xlabel('Categories')
ax.set_ylabel('Quantity')
plt.show()
```

```python
# histogram for a numeric variable
plot1=df['MPrice'].plot(kind='hist', 
                  bins = 10, 
                  title='Item Price')
plot1.set_xlabel('Price Range')
plt.show()
```

```python
# density plot
plot2=df['MPrice'].plot(kind='kde', 
                  title='Item Price')
plot2.set_xlabel('Price Range')
plt.show()
```

```python
plot2=sns.scatterplot(x='RQty', 
                y = 'TransTotal', 
                hue='Type',
                data=df)
plot2.set_title('quantity vs. transaction')
```

```python
colnames=df.columns
sns.pairplot(df, vars=[
                       'RPrice', 'MPrice', 
                       'MQty', 'LineTotal', 
                       'TransTotal'])
plt.show()
```

```python
# correlation table
df_corr=df[['RPrice', 'MPrice', 
                       'MQty', 'LineTotal', 
                       'TransTotal']].dropna().corr()
```

```python
# correlation heatmap
sns.heatmap(data=df_corr, annot=True)
```
